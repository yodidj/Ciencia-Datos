#### 1. Lectura archivo Datos_secop_I

```python
import pandas as pd

# URL del archivo CSV en formato "raw"
url1 = 'https://raw.githubusercontent.com/yodidj/Ciencia-Datos/main/datos_secop_I.csv'

# Leer el archivo CSV
df1 = pd.read_csv(url1)

# Mostrar las primeras filas del DataFrame
print(df1.head())
```

### Limpieza de los caracteres especiales y se unifica el NIT a 9 Digitos para df1 --> secop1 

```python
# Eliminar caracteres no numéricos y mantener solo los primeros 9 dígitos
df1['nit_entidad'] = df1['nit_entidad'].str.replace(r'\D', '', regex=True).str.slice(0, 9)

# Eliminar caracteres especiales no espacios columna "nombre_entidad"
df1['nombre_entidad'] = df1['nombre_entidad'].str.replace(r'[^A-Za-z0-9\s]', '', regex=True)


# Listar la columna 'NIT'
columna_nit = df1['nit_entidad']
nombre_ent = df1['nombre_entidad']
print(columna_nit)
print(nombre_ent)
```

```python
# Contar el número de filas del DataFrame 
num_filas = df1.shape[0]

print(f"El DataFrame concatenado tiene {num_filas} filas.")
```

#### 2. Lectura archivo Datos_secop_II

```python
import pandas as pd

# Leer el archivo CSV
df2 = pd.read_csv(r'https://raw.githubusercontent.com/yodidj/Ciencia-Datos/main/datos_secop_II.csv', sep = ';', encoding = 'latin-1')
print(df2.head())


```

### Limpieza de los caracteres especiales y se unifica el NIT a 9 Digitos para df2 --> secop2

```python
# Eliminar caracteres no numéricos y mantener solo los primeros 9 dígitos
df2['nit_entidad'] = df2['nit_entidad'].str.replace(r'\D', '', regex=True).str.slice(0, 9)

# Eliminar caracteres especiales no espacios columna "nombre_entidad"
df2['nombre_entidad'] = df2['nombre_entidad'].str.replace(r'[^A-Za-z0-9\s]', '', regex=True)


# Listar la columna 'NIT' y 'nombre entidad'
columna_nit = df2['nit_entidad']
print(columna_nit)
nombre_ent = df2['nombre_entidad']
print(columna_nit)
print(nombre_ent)

```

##### Se renombra df2 las columnas similares con la función 'rename()' con el fin de unificar el año, fecha de actualización y presupuesto para luego concatenar los dataframe df1 y df2
##### se usa la funcion 'concat' para unificar los registros en uno solo guardando el archivo final df_concatenado

```python
df2.rename(columns={
    'anno': 'anno_paa',
    'fecha_de_ultima_modificacion':'fecha_ultima_actualizacion',
    'valor_presupuesto_general':'ppto_total',
    
}, inplace=True)


# Concatenar los DataFrames
df_concatenado = pd.concat([df1, df2], ignore_index=True, sort=False)

# Mostrar las primeras filas del DataFrame concatenado
print(df_concatenado.head())
```

##### Finalmente se guarda el archivo con la función 'to_csv' y se cuentan cuantos registros quedaron 

```python
# Guardar el DataFrame concatenado en un archivo CSV
df_concatenado.to_csv('archivo_concatenado.csv', index=False)

# Contar el número de filas del DataFrame concatenado
num_filas = df_concatenado.shape[0]

print(f"El DataFrame concatenado tiene {num_filas} filas.")
```

#### Se puede filtrar por año en el siguiente caso 2024 

```python
# Filtrar filas donde 'anno_paa' es 2024
df_filtrado = df_concatenado[df_concatenado['anno_paa'] == 2024]

# Mostrar las primeras filas del DataFrame filtrado por el año 2024
print(df_filtrado.head())
```

```python
# Asegurarse de que la columna 'ppto_total' sea de tipo string
df_concatenado['ppto_total'] = df_concatenado['ppto_total'].astype(str)

# Eliminar caracteres especiales pero mantener los números y el punto decimal
df_concatenado['ppto_total'] = df_concatenado['ppto_total'].str.replace(r'[^\d.]', '', regex=True)

# Eliminar puntos que actúan como separadores de miles
df_concatenado['ppto_total'] = df_concatenado['ppto_total'].str.replace('.', '', regex=False)

# Rellenar valores vacíos con 0
df_concatenado['ppto_total'] = df_concatenado['ppto_total'].replace('', '0')

# Convertir la columna a tipo float
df_concatenado['ppto_total'] = df_concatenado['ppto_total'].astype(float)

#print(df_concatenado.head())
columna_ppto = df_concatenado['ppto_total']
print(columna_ppto)
```

### Se importa la libreria Matplotlib para poder generar los graficos despues de la instalación anterior

```python
import pandas as pd
import matplotlib.pyplot as plt

# Se toma el archivo concatenado para hacer ajuste en los campos
data = df_concatenado
df = pd.DataFrame(data)

# Se asegura que los datos que no son numeros coloque en el campo NaN
df['anno_paa'] = df['anno_paa'].astype(str)
df['ppto_total'] = pd.to_numeric(df['ppto_total'], errors='coerce')

if df['ppto_total'].isnull().any():
    print('Advertencia: Hay valores NaN en ppto_total después de la conversión. Estos serán eliminados.')
    df = df.dropna(subset=['ppto_total'])

# Dibujar la información de los datos dados
plt.figure(figsize=(10, 6))
plt.plot(df['anno_paa'], df['ppto_total'], marker='o', linestyle='-', color='b')

# Se colocan los titulos del grafico
plt.title('Presupuesto Total por Año')
plt.xlabel('Año')
plt.ylabel('Presupuesto Total MM')
plt.show()
```

### La siguente grafica podemos determinar el total de contratos por año tomando como dato la columna 'anno_paa'

```python
import pandas as pd
import matplotlib.pyplot as plt

# Se toma el archivo concatenado para hacer ajuste en los campos
data = df_concatenado
df = pd.DataFrame(data)

#La columna 'anno_paa' sea de tipo string
df['anno_paa'] = df['anno_paa'].astype(str)

# Contar la cantidad de contratos por año
contratos_por_anno = df['anno_paa'].value_counts().sort_index()

# Dibujar la información de los datos dados
plt.figure(figsize=(10, 6))
contratos_por_anno.plot(kind='bar', color='y')

# Títulos del gráfico
plt.title('Cantidad de Contratos por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Contratos')
plt.grid(True)
plt.show()
```
